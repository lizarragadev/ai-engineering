```python
import time

def static_llm_pipeline(prompt: str) -> str:
    return "Parece que estás pidiendo un resultado aritmético. Mi mejor estimación es 111."

def calculator(expr: str) -> float:
    allowed = set("0123456789.+-*/() ")
    if any(ch not in allowed for ch in expr):
        raise ValueError("La expresión contiene caracteres no permitidos.")
    return eval(expr)

def agent_answer(question: str) -> float:
    print(f"\n[Agente] Consulta recibida: {question}")
    time.sleep(0.2)

    print("[Pensamiento] Para responder esto, primero debo multiplicar 22 * 4.2 y luego sumar 19.")
    time.sleep(0.2)

    print("[Acción] calculator('22 * 4.2')")
    r1 = calculator("22 * 4.2")
    print(f"[Observación] 22 * 4.2 = {r1}")
    time.sleep(0.2)

    print("[Acción] calculator('92.4 + 19')")
    r2 = calculator("92.4 + 19")
    print(f"[Observación] 92.4 + 19 = {r2}")
    time.sleep(0.2)

    print(f"[Respuesta Final] El resultado es {r2}.")
    return r2


print("\n--- Ejemplo práctico (Código) ---")
print("Primero, así es como podría responder un pipeline LLM estático (sin herramientas, respuesta única):")
time.sleep(0.6)

baseline = static_llm_pipeline("¿Cuánto es 22 por 4.2 más 19?")
print("[Salida del LLM estático]:", baseline)
time.sleep(0.8)

print("\nAhora usemos un agente que piensa y llama herramientas paso a paso:")
print("""
Plan:
Paso 1: calcular 22 * 4.2
Paso 2: sumar 19 al resultado
Herramientas: calculator -> calculator
Resultado esperado: 111.4
""")
time.sleep(0.8)

agent_answer("¿Cuánto es 22 por 4.2 más 19?")
```
